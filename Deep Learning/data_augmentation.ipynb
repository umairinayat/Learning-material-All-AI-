```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
img = image.load_img('cat.8633.jpg',target_size=(200,200))
```

```python

import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
img = img.reshape((1,) + img.shape)
```

```python
i = 0
for batch in datagen.flow(img,save_to_dir='aug',batch_size=1,save_prefix='cat',save_format='jpeg'):
    i += 1
    if i > 20:
        break
```
